```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('Data/Rainbow.jpeg', 0) # Reading image as a gray scale
plt.imshow(img, cmap='gray')
```

Threshold -> set all the values to 0 below threshold and set to 1 when above the threshold

```python
img.max()
```

```python
threshold = 255/2
ret, thresholdImg = cv2.threshold(img, threshold, 255, cv2.THRESH_BINARY)
ret
```

```python
plt.imshow(thresholdImg, cmap='gray')
```

```python
ret, thresholdImgInv = cv2.threshold(img, threshold, 255, cv2.THRESH_BINARY_INV)
plt.imshow(thresholdImgInv, cmap='gray')
```

Real-time example: Enhancing an image using threshold concept

```python
crosswordImg = cv2.imread('Data/crossword_puzzle.png', 0)
plt.imshow(crosswordImg, cmap='gray')
```

```python
# Function to view image more readable
def Display_Image(img):
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
    
Display_Image(crosswordImg)
```

```python
ret, threshold_1 = cv2.threshold(crosswordImg, 127, 255, cv2.THRESH_BINARY) # 127 --> threshold value = max_val/2 = 255/2
Display_Image(threshold_1)
```

Adaptive Threshold method

```python
num_of_neighbor_blocks = 25
const_tuning_param = 8
threshold_2 = cv2.adaptiveThreshold(crosswordImg, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, num_of_neighbor_blocks, const_tuning_param)
Display_Image(threshold_2)
```

Let us Blend threshold_1 and threshold_2 for better results

```python
blended_img = cv2.addWeighted(threshold_1, 0.6, threshold_2, 0.4, gamma=0)
Display_Image(blended_img)
```
