```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import cv2
```

```python
# here we don't need to convert into numpy array. OpenCv does it.
img = cv2.imread('Data/Puppy.jpg') # Alert: Be cautious with the path. It will not give error if the path is wrong.
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
```

The above the image is giving bluish image due to different order of channels reading
# MATPLOTLIB --> RGB (Red Green Blue)
# OPENCV ------> BGR (Bue Green Red)

```python
convertedImg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #converting from BGR -> RGB
plt.imshow(convertedImg)
```

```python
img_gray = cv2.imread('Data/Puppy.jpg', cv2.IMREAD_GRAYSCALE)
img_gray.shape
```

```python
plt.imshow(img_gray) #it used default color mapping. Hence, it shows yellowish green scale
```

```python
plt.imshow(img_gray, cmap='gray')
```

Image resizing

```python
resizedImg = cv2.resize(convertedImg, (1000, 400))
plt.imshow(resizedImg)
```

```python
width_ratio = 0.5
height_ratio = 0.5
halfImg = cv2.resize(convertedImg, (0,0), convertedImg, width_ratio, height_ratio)
plt.imshow(halfImg)
```

```python
flippedImg = cv2.flip(convertedImg, 0)
plt.imshow(flippedImg)
```

Writing an image to a file

```python
cv2.imwrite('Data/outFlipped.jpg', flippedImg)
```
