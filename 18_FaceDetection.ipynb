```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
honorable_president = cv2.imread('Data/Dr_APJ_Kalam_Sir.jpeg', 0) # Reading image as gray scale
plt.imshow(honorable_president, cmap='gray')
```

Face cascade

```python
face_cascade = cv2.CascadeClassifier('Data/haar_cascades/haarcascade_frontalface_default.xml')
```

Function to detect faces

```python
def detect_face(img):
    face_img = img.copy()
    face_rect = face_cascade.detectMultiScale(face_img)
    
    # Create a rectangle over face
    for (x,y,w,h) in face_rect:
        cv2.rectangle(face_img, (x,y), (x+w, y+h), (255, 255, 255), 2)
    
    return face_img
```

```python
face = detect_face(honorable_president)
plt.imshow(face, cmap='gray')
```

```python
comedy_actor_vivek = cv2.imread('Data/Padmashree_Vivek_Actor.jpg', 0)
face_vivek = detect_face(comedy_actor_vivek)
plt.imshow(face_vivek, cmap='gray')
```

```python
scientist_group_pic = cv2.imread('Data/greatest_scientists.jpg', 0)
scientist_faces = detect_face(scientist_group_pic)
plt.imshow(scientist_faces, cmap='gray')
```

Function to detect face with scale factor for adjustment

```python
def detect_face_with_scaleFactor(img):
    face_img = img.copy()
    face_rect = face_cascade.detectMultiScale(face_img, scaleFactor=1.4, minNeighbors=1)
    # scale factor and minNeighbors -> experimental values. adjust it based on the performance.
    
    # Create a rectangle over face
    for (x,y,w,h) in face_rect:
        cv2.rectangle(face_img, (x,y), (x+w, y+h), (255, 255, 255), 2)
    
    return face_img
```

```python
scientist_faces = detect_face_with_scaleFactor(scientist_group_pic)
plt.imshow(scientist_faces, cmap='gray')
```

Eye Cascade

```python
eye_cascade = cv2.CascadeClassifier('Data/haar_cascades/haarcascade_eye.xml')

def detect_eyes(img):
    eye_img = img.copy()
    eye_rect = eye_cascade.detectMultiScale(eye_img)
    
    # Create a rectangle over face
    for (x,y,w,h) in eye_rect:
        cv2.rectangle(eye_img, (x,y), (x+w, y+h), (255, 255, 255), 2)
    
    return eye_img
```

```python
vivek = detect_eyes(comedy_actor_vivek)
plt.imshow(vivek, cmap='gray')
```

Adjust the multiscale for better results

```python
def detect_eyes_adjustment(img):
    eye_img = img.copy()
    eye_rect = eye_cascade.detectMultiScale(eye_img, scaleFactor=4.1, minNeighbors=3)
    
    # Create a rectangle over face
    for (x,y,w,h) in eye_rect:
        cv2.rectangle(eye_img, (x,y), (x+w, y+h), (255, 255, 255), 2)
    
    return eye_img

vivek = detect_eyes_adjustment(comedy_actor_vivek)
plt.imshow(vivek, cmap='gray')
```

Note: For better results, use unedited image
