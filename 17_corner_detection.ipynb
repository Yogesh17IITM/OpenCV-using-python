```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Harris Corner detection

```python
img = cv2.imread('Data/Chessboard.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
```

```python
gray
```

```python
gray = np.float32(gray) # convert all the gray chess values into float
dst = cv2.cornerHarris(gray, blockSize=3, ksize=3, k=0.04)
plt.imshow(dst, cmap='gray')
```

```python
dst = cv2.dilate(dst, None)
plt.imshow(dst)
```

```python
img[dst>0.01*dst.max()] = [255, 0 , 0]
plt.imshow(img)
```

```python
realChessImg = cv2.imread('Data/RealChessBoard.jpeg')
realChessImg = cv2.cvtColor(realChessImg, cv2.COLOR_BGR2RGB)
gray_realchess = cv2.cvtColor(realChessImg, cv2.COLOR_RGB2GRAY)
gray_realchess = np.float32(gray_realchess)
dst = cv2.cornerHarris(src=gray_realchess, blockSize=10, ksize=1, k=0.04)
dst = cv2.dilate(dst, None)
realChessImg[dst>0.01*dst.max()] = [255, 0, 0]
plt.imshow(realChessImg)
```

Sgi-Tomasi Corner detection Algorithm

```python
flatchess = cv2.imread('Data/Chessboard.png')
flatchess = cv2.cvtColor(flatchess, cv2.COLOR_BGR2RGB)
gray_chess = cv2.cvtColor(flatchess, cv2.COLOR_RGB2GRAY)

corners = cv2.goodFeaturesToTrack(gray_chess,64, 0.01, 10) # arg@2 -> no. of corners to track
corners = np.int32(corners)
for i in corners:
    x,y = i.ravel()
    cv2.circle(flatchess, (x,y), 3, (255,0,0), -1)

plt.imshow(flatchess)
```
