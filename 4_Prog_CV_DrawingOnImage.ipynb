```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(500,500,3), dtype=np.int16)
plt.imshow(blank_img)
```

Draw rectangles on the image

```python
cv2.rectangle(blank_img, pt1=(300,10), pt2=(450,200), color=(255,0,0), thickness=5)
plt.imshow(blank_img)
```

Draw a filled shaped on the image

```python
cv2.circle(img=blank_img, center=(250,250), radius=50, color=(0,255,0), thickness=-1)
plt.imshow(blank_img)
```

Writing text on the image

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text='Warning', org=(10,400), fontFace=font, fontScale=2, color=(255,0,0), thickness=3, lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```
