```python
import numpy as np
import matplotlib.pyplot as plt

#allows to plot within notebook
%matplotlib inline
```

```python
# import python image library
from PIL import Image
pic = Image.open('Data/Puppy.jpg')
```

```python
type(pic)
```

Convert image into array and check the shape (Height, Width, No. of channels)

```python
pic_array = np.asarray(pic)
pic_array.shape
```

```python
plt.imshow(pic_array)
```

Display only one channel from a 3 channel image [R, G, B]

```python
pic_red = pic_array.copy()
plt.imshow(pic_red[:,:,0])
```

```python
# Display only Red Channel values
pic_red[:,:,0]
```

Grayscale image

```python
plt.imshow(pic_array[:,:,0], cmap='gray')
```

```python
plt.imshow(pic_array[:,:,1], cmap='gray')
```

Set value to 0 for G,B color channels and Keep only Red channel

```python
pic_red[:,:,1:2] = 0
plt.imshow(pic_red)
```
